```python
import pandas as pd
from sklearn import datasets
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
base = pd.read_csv(r"C:\Users\dylan\Documents\GitHub\Consumo-Alcool-Russia\Consumption of alcoholic beverages 2017-2023 (Pivot table).csv", encoding='latin1')
```

## Evolução do consumo por ano

```python
base_ajustada = base[['Year','Consumption of alcoholic beverages (in liters per capita)']]
base_ajustada = base_ajustada.groupby(by='Year').mean()
```

```python
# Gráfico
x = base_ajustada.index
y = base_ajustada[base_ajustada.columns[0]]

plt.figure(figsize=(10, 6))
plt.plot(base_ajustada.index, base_ajustada['Consumption of alcoholic beverages (in liters per capita)'], marker='o')

plt.title('Consumo de Bebidas Alcoólicas por Ano', fontsize=16)
plt.xlabel('Ano', fontsize=12)
plt.ylabel('Consumo (litros)', fontsize=12)
plt.grid(visible=True, linestyle='--', alpha=0.5)

plt.show()
```

```python
# Tabela
base_ajustada["Aumento (%)"] = base_ajustada["Consumption of alcoholic beverages (in liters per capita)"].pct_change() * 100
base_ajustada["Aumento (%)"] = base_ajustada["Aumento (%)"].fillna("-")

for numerador,valor in enumerate(base_ajustada['Aumento (%)']):
    if valor != '-':
        base_ajustada['Aumento (%)'].iloc[numerador] = f'{base_ajustada['Aumento (%)'].iloc[numerador]:.2f}'

base_ajustada
```
