```python
import pandas as pd
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
base = pd.read_csv(r"C:\Users\dylan\Documents\GitHub\scikit learn - Previsão de bebâdo\Consumption of alcoholic beverages 2017-2023 (Pivot table).csv", encoding='latin1')
```

```python
base_ajustada = base[['Year','Consumption of alcoholic beverages (in liters per capita)']]
base_ajustada = base_ajustada.groupby(by='Year').mean()
```

```python
x = base_ajustada.index
y = base_ajustada[base_ajustada.columns[0]]
```

```python
# Criar o gráfico de linha
plt.figure(figsize=(10, 6))
plt.plot(base_ajustada.index, base_ajustada['Consumption of alcoholic beverages (in liters per capita)'], marker='o', linestyle='-', color='b')

# Adicionar títulos e rótulos
plt.title('Consumo de Bebidas Alcoólicas por Ano', fontsize=14)
plt.xlabel('Ano', fontsize=12)
plt.ylabel('Consumo (litros per capita)', fontsize=12)

# Exibir grades no gráfico
plt.grid(visible=True, linestyle='--', alpha=0.5)

# Ajustar limites (opcional)
#plt.xticks(base_ajustada.index, rotation=0)  # Rotacionar anos se necessário
#plt.tight_layout()

# Exibir o gráfico
plt.show()
```
